# 8. Removing Unnecessary Features

`customerID` sütunu her müşteri için benzersiz bir kimliktir.
Makine öğrenmesi modeli için tahmin gücü taşımadığı için veri setinden çıkarılacaktır.

```python
df.drop("customerID", axis=1, inplace=True)

df.head()
```

# 9. Detecting Hidden Missing Values

Bu veri setinde eksik değerler klasik NULL değildir.
Bazı kayıtlar boş karakter (" ") içerdiğinden ayrıca kontrol edilmiştir.

```python
(df["TotalCharges"] == " ").sum()
```

# 10. Converting TotalCharges to Numeric

`TotalCharges` sütunu metin tipindedir.
Makine öğrenmesi algoritmalarında kullanılabilmesi için sayısal tipe dönüştürülmüştür.

```python
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors="coerce")
```

# 11. Handling Missing Values

Eksik kayıt sayısı toplam veri setine göre oldukça düşüktür.
Bu nedenle eksik kayıtlar veri setinden çıkarılmıştır.

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

## Observation

Toplam 11 kayıt `TotalCharges` sütununda boş karakter içerdiği için
sayısal dönüşüm sırasında eksik değere dönüşmüştür.

Eksik veri oranı %1'in altında olduğundan kayıtlar silinmiştir.

Bu işlem veri kaybını minimum seviyede tutarken model kalitesini artıracaktır.
