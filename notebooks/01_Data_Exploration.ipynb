# 15. Feature Engineering and Data Preparation

```python
# Orijinal veriyi korumak için kopya oluşturuyoruz
df_model = df.copy()

df_model.head()
```

```python
df_model["Churn"] = df_model["Churn"].map({
    "Yes": 1,
    "No": 0
})

df_model["Churn"].value_counts()
```

```python
categorical_columns = df_model.select_dtypes(include="object").columns.tolist()

numerical_columns = df_model.select_dtypes(include=["int64", "float64"]).columns.tolist()

print("Kategorik sütunlar:")
print(categorical_columns)

print("\nSayısal sütunlar:")
print(numerical_columns)
```

```python
df_encoded = pd.get_dummies(
    df_model,
    columns=categorical_columns,
    drop_first=True
)

df_encoded.head()
```

```python
print(df_encoded.shape)
```

```python
X = df_encoded.drop("Churn", axis=1)

y = df_encoded["Churn"]

print(X.shape)
print(y.shape)
```

## Train-Test Split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)
```

```python
print("Train:", X_train.shape)

print("Test :", X_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)

X_test_scaled = scaler.transform(X_test)
```

```python
print(X_train_scaled.shape)

print(X_test_scaled.shape)
```

# 16. Model 1 - Logistic Regression


```python
from sklearn.linear_model import LogisticRegression

log_model = LogisticRegression(
    random_state=42,
    max_iter=1000
)

log_model.fit(
    X_train_scaled,
    y_train
)
```

```python
y_pred_log = log_model.predict(X_test_scaled)

y_prob_log = log_model.predict_proba(X_test_scaled)[:,1]
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(
    y_test,
    y_pred_log
)
```

```python
from sklearn.metrics import (
    precision_score,
    recall_score,
    f1_score
)

print("Accuracy :", accuracy_score(y_test, y_pred_log))
print("Precision:", precision_score(y_test, y_pred_log))
print("Recall   :", recall_score(y_test, y_pred_log))
print("F1 Score :", f1_score(y_test, y_pred_log))
```

```python
from sklearn.metrics import roc_auc_score

roc_auc_score(
    y_test,
    y_prob_log
)
```

```python
from sklearn.metrics import classification_report

print(
    classification_report(
        y_test,
        y_pred_log
    )
)
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(
    y_test,
    y_pred_log
)

print(cm)
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import ConfusionMatrixDisplay

disp = ConfusionMatrixDisplay(
    confusion_matrix=cm
)

disp.plot()

plt.show()
```

### Model Evaluation

Logistic Regression modeli yaklaşık %80 doğruluk oranı elde etmiştir.

Ancak veri seti tam dengeli olmadığı için yalnızca Accuracy metriği değerlendirilmemiştir.

Precision, Recall ve F1 Score değerleri de incelenerek model performansı analiz edilmiştir.

# 17. Model 2 - Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_model = RandomForestClassifier(
    n_estimators=200,
    random_state=42
)

rf_model.fit(
    X_train,
    y_train
)
```

```python
y_pred_rf = rf_model.predict(X_test)

y_prob_rf = rf_model.predict_proba(X_test)[:,1]
```

```python
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score
)

print("Accuracy :", accuracy_score(y_test, y_pred_rf))
print("Precision:", precision_score(y_test, y_pred_rf))
print("Recall   :", recall_score(y_test, y_pred_rf))
print("F1 Score :", f1_score(y_test, y_pred_rf))
print("ROC AUC  :", roc_auc_score(y_test, y_prob_rf))
```

```python
from sklearn.metrics import classification_report

print(
    classification_report(
        y_test,
        y_pred_rf
    )
)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

cm_rf = confusion_matrix(
    y_test,
    y_pred_rf
)

disp = ConfusionMatrixDisplay(
    confusion_matrix=cm_rf
)

disp.plot()

plt.show()
```

### Model Evaluation

Random Forest modeli doğrusal olmayan ilişkileri de öğrenebildiği için
Logistic Regression modeline göre farklı bir performans göstermiştir.

Bu model daha sonra diğer model ile karşılaştırılacaktır.

```python
comparison = pd.DataFrame({
    "Model": [
        "Logistic Regression",
        "Random Forest"
    ],

    "Accuracy": [
        accuracy_score(y_test, y_pred_log),
        accuracy_score(y_test, y_pred_rf)
    ],

    "Precision": [
        precision_score(y_test, y_pred_log),
        precision_score(y_test, y_pred_rf)
    ],

    "Recall": [
        recall_score(y_test, y_pred_log),
        recall_score(y_test, y_pred_rf)
    ],

    "F1 Score": [
        f1_score(y_test, y_pred_log),
        f1_score(y_test, y_pred_rf)
    ],

    "ROC-AUC": [
        roc_auc_score(y_test, y_prob_log),
        roc_auc_score(y_test, y_prob_rf)
    ]
})

comparison
```
